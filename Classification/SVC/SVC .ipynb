```python
#importing the libraries
import pandas as pd # data processing I\O CSV
import numpy as np #linear algebra
import matplotlib.pyplot as plt
#machine learniung module
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python

#import datasets
path='heart.csv'
dataset=pd.read_csv(path)
dataset.head(10)
```

```python
#we can describtion dataset
dataset.describe()
```

```python
#in this lession we can split dataset
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1:]
print(X.shape)
print(y.shape)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,shuffle=True)
```

# support vector classifier model (SVC)

```python
kernel = ['linear','poly','rbf','sigmoid','precomputed']
#if we used poly we must enter the degree
SVRModel=SVC(kernel='linear',max_iter=10000,C=0.1)
SVRModel.fit(X_train,y_train)
```

```python
#we can calculation prdict
y_pred=SVRModel.predict(X_test)
#Calculation score
print('SVRModel score train : ',SVRModel.score(X_train,y_train)*100)
print('SVRModel score test : ',accuracy_score(y_pred,y_test)*100)
```

```python
#importing the libraries
import pandas as pd # data processing I\O CSV
import numpy as np #linear algebra
import matplotlib.pyplot as plt
#machine learniung module
from sklearn.linear_model import LogisticRegression

import warnings
warnings.filterwarnings('ignore')
```

```python
#import datasets
path='heart.csv'
dataset=pd.read_csv(path)
dataset.head(10)
```

```python
#we can describtion dataset
dataset.describe()
```

```python
#in this lession we can split dataset
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1:]
print(X.shape)
print(y.shape)
```

solver=[‘newton-cg’, ‘lbfgs’, ‘liblinear’, ‘sag’, ‘saga’], default=’lbfgs’
Algorithm to use in the optimization problem.

For small datasets, ‘liblinear’ is a good choice, whereas ‘sag’ and ‘saga’ are faster for large ones.

For multiclass problems, only ‘newton-cg’, ‘sag’, ‘saga’ and ‘lbfgs’ handle multinomial loss; ‘liblinear’ is limited to one-versus-rest schemes.

‘newton-cg’, ‘lbfgs’, ‘sag’ and ‘saga’ handle L2 or no penalty

‘liblinear’ and ‘saga’ also handle L1 penalty

‘saga’ also supports ‘elasticnet’ penalty

‘liblinear’ does not support setting penalty='none'

C : Inverse of regularization strength; must be a positive float. Like in support vector machines, smaller values specify stronger regularization.

```python
Log_reg_model=LogisticRegression(solver='liblinear',penalty='l2',C=1,max_iter=100,tol=1e-4)
#now we can fit dataset
Log_reg_model.fit(X,y)
```

```python
#we can calculation prdict
y_pred=Log_reg_model.predict(X)
#Calculation score
print('LogisticRegression score train : ',Log_reg_model.score(X,y))
```
